```python
q=35000
p=15000
train=train_df.head(q)
test=test_df.head(p)
```

```python
X_train = train.drop(["TrialID","ObsNum","IsAlert"], axis=1)
Y_train = train["IsAlert"]
X_test  = test.drop(["TrialID","ObsNum","IsAlert"], axis=1).copy()
#_test  = test_df.drop(["TrialID","ObsNum"], axis=1)
X_train.shape, Y_train.shape, X_test.shape
```

```python
from sklearn.preprocessing import StandardScaler
sc = StandardScaler()
X_train= sc.fit_transform(X_train) 
X_test = sc.transform(X_test)  
```

```python
logreg = LogisticRegression()
logreg.fit(X_train, Y_train)
Y_pred = logreg.predict(X_test)
acc_log = round(logreg.score(X_train, Y_train) * 100, 2)
print acc_log
print(classification_report(solution_df['Prediction'].head(p), Y_pred))
print(confusion_matrix(solution_df['Prediction'].head(p), Y_pred))
```

```python
coeff_df = pd.DataFrame(train_df.columns.delete([0,1,2]))
coeff_df.columns = ['Feature']
coeff_df["Correlation"] = pd.Series(logreg.coef_[0])
coeff_df.sort_values(by='Correlation', ascending=False)
```

```python
svc = SVC()
svc.fit(X_train, Y_train)
Y_pred = svc.predict(X_test)
acc_svc = round(svc.score(X_train, Y_train) * 100, 2)
print acc_svc
print(classification_report(solution_df['Prediction'].head(p), Y_pred))
print(confusion_matrix(solution_df['Prediction'].head(p), Y_pred))
```
